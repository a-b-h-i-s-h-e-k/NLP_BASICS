# Naive Bayes: Predicting survival from titanic crash

- P(A/B) = P(B/A) * P(A) / P(B) -> Conditional Probability <br>

- Bayes theroem -> P (Survived / Male & Class & Age & Cabin & Fare)<br>

- I am supposing/ or making a 'Naive' assumptio that featrues such as male, class, age, cabin, fare etc. are independent of each other <br>

```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
df
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

i drop these columns becoz for me now they are not relevant to learn anive byas

```python
inputs = df.drop('Survived',axis='columns')
target = df.Survived
```

```python
#inputs.Sex = inputs.Sex.map({'male': 1, 'female': 2})
```

```python
dummies = pd.get_dummies(inputs.Sex)
dummies.head(3)
```

```python
inputs = pd.concat([inputs,dummies],axis='columns')
inputs.head(3)
```

I am dropping male column as well because of dummy variable trap theory. One column is enough to repressent male vs female

```python
inputs.drop(['Sex','male'],axis='columns',inplace=True)
inputs.head(3)
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age[:10]
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)

# we use naive gussian method when data is in normal distribution
```

```python
model.score(X_test,y_test)
```
